```python
from ultralytics import YOLO
model = YOLO('/content/runs/detect/train/weights/best.pt')
```

```python
import os

# ==============================================================================
# ⚠️ 1. USER CONFIGURATION: UPDATE THESE VARIABLES
# ==============================================================================

# Path to your trained model weights (e.g., 'runs/detect/train/weights/best.pt')
MODEL_PATH = '/content/runs/detect/train/weights/best.pt'

# Path to the specific image you want to test
IMAGE_TO_TEST_PATH = '/content/continuous_fire-6/test/images/------2022-05-16-201833_png.rf.67a6414cf36e8a74eb3ba01a24b7f5cf.jpg'
# Directory where the results (image with boxes) will be saved
# A new folder (e.g., 'runs/detect/predict') will be created here.
SAVE_DIR = 'runs/predict'

# ==============================================================================
# 2. RUN INFERENCE
# ==============================================================================

print(f"Loading model from: {MODEL_PATH}...")
try:
    # Load the YOLO model
    model = YOLO(MODEL_PATH)
    print("Model loaded successfully.")

except Exception as e:
    print(f"\n❌ Error loading model: {e}")
    print("Ensure you have 'ultralytics' installed (pip install ultralytics)")
    exit()

print(f"\nRunning prediction on: {IMAGE_TO_TEST_PATH}...")

# Run prediction on the image
# 'source': The path to the image
# 'save': True means it saves the image with the bounding boxes drawn on it
# 'project': The root directory to save results (e.g., 'runs')
# 'name': The name of the folder inside the project (e.g., 'predict')
results = model.predict(
    source=IMAGE_TO_TEST_PATH,
    save=True,
    project=SAVE_DIR,
    name='single_test', # This will create 'runs/predict/single_test'
    exist_ok=True, # Overwrite previous runs in the same folder if it exists
    conf=0.25 # Minimum confidence threshold (adjust if needed)
)

# ==============================================================================
# 3. REPORT RESULTS LOCATION
# ==============================================================================

# The resulting image will be saved inside a specific folder.
# This finds the exact path of the saved image.
# The filename will be the same as the input image.
result_folder = os.path.join(SAVE_DIR, 'single_test')
saved_image_path = os.path.join(result_folder, os.path.basename(IMAGE_TO_TEST_PATH))

print("\n--- ✅ INFERENCE COMPLETE ---")
print(f"   Prediction Run Folder: **{result_folder}**")
print(f"   Image with Detections Saved To: **{saved_image_path}**")

# You can also print the detected objects/classes and confidences
if results and results[0].boxes.data.shape[0] > 0:
    print("\n   Detected Objects:")
    for box in results[0].boxes:
        conf = box.conf.item() * 100
        cls_id = int(box.cls.item())
        cls_name = model.names[cls_id]
        print(f"     - Class: {cls_name}, Confidence: {conf:.2f}%")
else:
    print("   No objects detected above the confidence threshold.")
```
